** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt

# Create a figure object
fig = plt.figure()

# Create subplots with 1 row and 2 columns
f1, ax = plt.subplots(ncols=2)

# Plot (x, y) on the first subplot (left subplot)
x = np.arange(0,100)
y = x*2
z = x**2
ax[0].plot(x, y,linestyle='--')
ax[0].set_title('Left Subplot')  # Set title for the left subplot

# Plot (x, y) on the second subplot (right subplot)
ax[1].plot(x, z,color="red")
ax[1].set_title('Right Subplot')  # Set title for the right subplot

# Adjust layout to prevent overlap of titles
plt.tight_layout()

plt.show()

```
