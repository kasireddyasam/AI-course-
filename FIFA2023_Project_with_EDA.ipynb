```python
needed_cols=['ID', 'Name', 'FullName', 'Age', 'Height', 'Weight' ,
        'Overall', 'Potential', 'Growth', 'TotalStats',
       'BaseStats', 'BestPosition', 'Club', 'ValueEUR', 'WageEUR',
       'ReleaseClause', 'ContractUntil',
       'ClubJoined', 'OnLoad',  'PreferredFoot', 'IntReputation', 'WeakFoot','Nationality',
       'SkillMoves', 'AttackingWorkRate', 'DefensiveWorkRate', 'PaceTotal',
       'ShootingTotal', 'PassingTotal', 'DribblingTotal', 'DefendingTotal',
       'PhysicalityTotal', 'Crossing', 'Finishing', 'HeadingAccuracy',
       'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
       'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength',
       'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle',
       'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']
```

#Does the Age of the player affect on his BallControl performance?


```python
 sns.displot(data['Age'])
 plt.axvline(np.mean(data['Age']),color='red',linestyle='--')
```

```python
sns.displot(data['BallControl'])
plt.axvline(np.mean(data['BallControl']),color='red',linestyle='--')
```

###.lmplot is use for creating linear regressionplot and show best fit line
* order = degree

```python
sns.lmplot(x='Age',y='BallControl',data=data,markers='*',order=2)
```

##2. Show how height affect different factors like stamina,dribbling,pace,passing and heading accuracy?


```python
fig=plt.gcf()
fig.set_size_inches(10,10)
sns.lineplot(x='Height',y='Stamina',data=data)
sns.lineplot(x='Height',y='Dribbling',data=data)
sns.lineplot(x='Height',y='PaceTotal',data=data)
sns.lineplot(x='Height',y='PassingTotal',data=data)
sns.lineplot(x='Height',y='HeadingAccuracy',data=data)
fig.legend(labels=['Stamina','Dribbling','PaceTotal','passingTotal','HeadingAccuracy'])
```

 Show if there is a relation

```python
sns.relplot(x='WageEUR', y='Overall', data=data)
```

Show the top Quickest players

```python
top_quicker=data.sort_values(by='Acceleration',ascending=False)
```

```python
sns.barplot(x='Name', y='Acceleration',data=top_quicker.head(20),palette='rainbow')
plt.xticks(rotation=90);
```

## 5.Determine if their is a relation b/w the best position amd Weage

```python
sns.scatterplot(x=data['BestPosition'],y=data['WageEUR'],data=data)
```

## Nationalty vs Wages


```python
top_wages=data.sort_values(by=['WageEUR'],ascending=False)
```

```python
top_wages.head()
```

```python
sns.countplot( y='Nationality',data=top_wages.head(100),palette=sns.color_palette('rainbow'))
plt.title('Nationality vs Wages');
```
